# Teorema central del límite

El Teorema del Límite Central (TLC) es fundamental en las áreas de Machine Learning (ML) y Deep Learning (DL), ya que proporciona una base teórica sólida para muchas técnicas estadísticas y de modelado de datos. A continuación, se presentan algunas aplicaciones específicas del TLC en estos campos:

+ Justificación para la inferencia estadística
+ Entrenamiento de redes neuronales y optimziación
+ Validación de modelos y evaluación de rendimiento
+ Reducción de dimensionalidad y preprocesamiento de datos

## Justificación para la inferencia estadística

El ***Teorema Central del límite*** establece que, dada una muestra suficientemente grande, la distribución de la media muestral se aproxima a una distribución normal, independientemente de la distribución original de los datos. Esto tiene implicaciones importantes en Machine Learning y Deep Learning:

+ Modelos de Regresión y Clasificación: En la mayoría de los algoritmos de aprendizaje supervisado, como la regresión lineal, logística y otros métodos basados en gradiente, se supone que los errores (residuos) tienen una distribución normal. El teorema cenrtal del límite proporciona una justificación teórica para esta suposición cuando el tamaño de la muestra es grande.

+ Intervalos de Confianza y Pruebas de Hipótesis: En Machine Learning y Deep Learning, se realizan pruebas estadísticas para validar los modelos, y el teorema central del límite permite que los intervalos de confianza y las pruebas de hipótesis sean válidas para la mayoría de las estadísticas.

## Entrenamiento de redes neuronales y optimización

+ Inicialización de Pesos: al entrenar redes neuronales profundas, los pesos se inicializan típicamente utilizando distribuciones normales. El teorema central del límite asegura que, con un gran número de nodos y conexiones, las sumas de muchas variables aleatorias (como activaciones) tienden a ser normalmente distribuidas. Esto facilita el cálculo de las actualizaciones de los pesos utilizando métodos basados en gradientes.

+ Propagación de Errores: en el entrenamiento de redes neuronales, los gradientes calculados a través del algoritmo de retropropagación a menudo se comportan como variables aleatorias. El teorema central del límite sugiere que la distribución de estos gradientes se aproximará a una distribución normal a medida que el tamaño de la red y la cantidad de muestras de entrenamiento aumentan, lo que estabiliza el proceso de optimización

## Validación de modelos y evaluación de rendimiento

+ Bootstraping y Métodos de Muestreo: el teorema central del límite se utiliza para justificar el uso de técnicas de remuestreo como el bootstrapping en Machine Learning. Estas técnicas permiten estimar el error y la varianza de los modelos, proporcionando distribuciones normales aproximadas de los estimadores cuando se remuestrean múltiples veces.

+ Curvas ROC y AUC: en la evaluación del rendimiento del modelo, las áreas bajo las curvas ROC (AUC) a menudo se aproximan a una distribución normal debido al teorema central del límite, lo que permite comparar diferentes modelos estadísticamente.

## Reducción de dimensionalidad y preprocesamiento de datos

+ PCA (Análisis de Componentes Principales): cuando se reduce la dimensionalidad de los datos mediante técnicas como PCA, el teorema central del límite asegura que, para un gran número de muestras, la suma de los componentes principales tiende a una distribución normal. Esto ayuda en la selección y retención de características más relevantes y facilita el análisis y la visualización.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configuración de estilo para seaborn
sns.set(style="whitegrid")

# Paso 1: Generar datos a partir de una distribución exponencial
np.random.seed(42)
data = np.random.exponential(scale=2, size=10000)

# Visualizar la distribución original
plt.figure(figsize=(10, 5))
sns.histplot(data, bins=30, kde=True, color='skyblue')
plt.title("Distribución Exponencial Original")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()

# Paso 2: Tomar muestras y calcular sus medias
sample_sizes = [5, 10, 30, 50, 100]  # Diferentes tamaños de muestra
plt.figure(figsize=(15, 10))

for i, size in enumerate(sample_sizes):
    sample_means = [np.mean(np.random.choice(data, size=size)) for _ in range(1000)]

    # Visualización de las medias muestrales
    plt.subplot(3, 2, i+1)
    sns.histplot(sample_means, bins=30, kde=True, color='coral')
    plt.title(f"Distribución de Medias Muestrales (n={size})")
    plt.xlabel("Media Muestral")
    plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```
