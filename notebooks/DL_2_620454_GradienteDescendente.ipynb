#Gradiente descendente

Es un algoritmo de optimización utilizado para encontrar el mínimo (o máximo) de una función. Es bastante usado en machine learning y redes neuronales para ***ajustar los parámetros del modelo y minimizar la función de pérdida***.

En términos simples, el gradiente descendente busca calcular el valor mínimo de una función matemática.

En este caso se va a explicar con la función $f(x) = x^2 + 1$, el gradiente es también conocido como la derivada, por tanto, se va a trabajar con la derivada: en este caso es $f(x)=2x$

El objetivo es encontrar el valor de $x$ que minimice la función $f(x) = x^2 + 1$. El algoritmo de gradiente descendente trabaja con:

+ Inicialización: elegir el valor inicial para $x$
+ Tasa de apendizaje (learning rate): se define una tasa que aprendizaje que controla el tamaño de los pasos que se dan en cada iteración del algoritmo. Se trata de un $hiper parámetro$ crítico que debe ajustarse cuidadosamente, ya que un valor muy pequeño puede hacer que el algoritmo sea lento y uno muy grande puede hacer que el algoritmo no converja correctamente.
+ Iteraciones: repeticiones donde se actualiza el valor de $x$ usando el gradiente (derivada) de la función y la tasa de aprendizaje.

`````python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
from numpy.random import randint
`````

`````python
def f(x):
  return x**2 + 1

def gradiente(x):
  return 2*x

def gradient_descent(learning_rate, epoch):
  x = 0 # Este valor inicial se define aleatoriamente por el método del gradiente descendente
  for _ in range(epoch):
    gradient = gradiente(x)
    x = x - learning_rate*gradient
  return x

# Hiper parámetros
learning_rate = 0.1
epochs = 100

# Ejecuta el gradiente descendente
result = gradient_descent(learning_rate, epochs)

print("El valor de x que minimiza la función es", result)
print("El valor de mínimo de la función es", f(result))
`````

`````python
# Mejora un poco la función para que entregue más información
def gradient_descent_best(learning_rate : float, epoch : int, x_init : int = 5 ):
  iteraciones = []
  y = []
  x = x_init # Este valor inicial se define aleatoriamente por el método del gradiente descendente
  for i in range(epoch):
    gradient = gradiente(x)
    x = x - learning_rate*gradient

    y.append(f(x))
    iteraciones.append(i+1)
  return x, iteraciones, y

# Hiper parámetros
learning_rate = 0.1
epochs = 100

# Ejecuta el gradiente descendente
result, iterations, y = gradient_descent_best(learning_rate, epochs)

print("El valor de x que minimiza la función es", result)
print("El valor de mínimo de la función es", f(result))
`````

`````python
plt.subplot(1,2,1)
plt.plot(iterations, y)
plt.xlabel("Iteraciones", fontsize=14, fontweight="bold")
plt.ylabel("y", fontsize=14, fontweight="bold")

X = np.linspace(-5,5,100)
Y = list(map(lambda x: f(x), X))

# Graficamos
plt.subplot(1,2,2)
plt.plot(X,Y,result,f(result),'ro')
plt.xlabel('x', fontsize=14, fontweight="bold")
plt.ylabel('y', fontsize=14, fontweight="bold")
plt.suptitle("Iteraciones y f(x)", fontsize=16, fontweight="bold")
plt.show()
`````

`````python
# Prueba con otra función
def fx(x):
  return x**2 -4*x + 4

def gradientex(x):
  return 2*x - 4
`````

`````python
def gradient_descent_best_X(learning_rate : float, epoch : int, x_init : int = 5 ):
  iteraciones = []
  y = []
  x = x_init # Este valor inicial se define aleatoriamente por el método del gradiente descendente
  for i in range(epoch):
    gradient = gradientex(x)
    x = x - learning_rate*gradient

    y.append(fx(x))
    iteraciones.append(i+1)
  return x, iteraciones, y

# Hiper parámetros
learning_rate = 0.1
epochs = 100

# Ejecuta el gradiente descendente
result, iterations, y = gradient_descent_best_X(learning_rate, epochs)

print("El valor de x que minimiza la función es", result)
print("El valor de mínimo de la función es", fx(result))
`````

`````python
plt.subplot(1,2,1)
plt.plot(iterations, y)
plt.xlabel("Iteraciones", fontsize=14, fontweight="bold")
plt.ylabel("y", fontsize=14, fontweight="bold")

X = np.linspace(-5,5,100)
Y = list(map(lambda x: fx(x), X))

# Graficamos
plt.subplot(1,2,2)
plt.plot(X,Y,result,fx(result),'ro')
plt.xlabel('x', fontsize=14, fontweight="bold")
plt.ylabel('y', fontsize=14, fontweight="bold")
plt.suptitle("Iteraciones y f(x)", fontsize=16, fontweight="bold")
plt.show()
`````

En este caso, el resultado será aproximadamente $x=2.0$, que es el valor que minimiza la función cuadrática $f(x)=x^2-4x+4$. La función tiene un mínimo global en $x = 2$ (la derivada es igual a cero cuando $x$ toma ese valor). El algoritmo pudo encontrar ese mínimo aproximadamente después de 100 iteraciones con una tasa de aprendizaje adecuada.

Existen otras técnicas para problemas más complejos: gradiente estocástico y mini-batch

Observando el gráfico de la izquierda es posible apreciar que desde la iteración 25 en adelante el cambio es casi insignificante por lo que veremos qué pasa si se disminuye la cantidad de iteraciones: ¿encontrará el mínimo global?

`````python
# Hiper parámetros
learning_rate = 0.1
epochs = 20

# Ejecuta el gradiente descendente
result, iterations, y = gradient_descent_best_X(learning_rate, epochs)

print("El valor de x que minimiza la función es", result)
print("El valor de mínimo de la función es", fx(result))

# Graficamos
plt.subplot(1,2,1)
plt.plot(iterations, y)
plt.xlabel("Iteraciones",fontsize=16, fontweight="bold")
plt.ylabel("y",fontsize=16, fontweight="bold")

X = np.linspace(-5,5,100)
Y = list(map(lambda x: fx(x), X))

# Graficamos
plt.subplot(1,2,2)
plt.plot(X,Y,result,fx(result),'ro')
plt.xlabel('x',fontsize=16, fontweight="bold")
plt.ylabel('y',fontsize=16, fontweight="bold")
plt.suptitle("Iteraciones y f(x)", fontsize=16, fontweight="bold")
plt.show()
`````

`````python
# Ahora aleatoriza el valor inicial de x y cambia la cantidad de épocas

x_inicial = randint(10)

# Hiper parámetros
learning_rate = 0.1
epochs = 15

# Ejecuta el gradiente descendente
result, iterations, y = gradient_descent_best_X(learning_rate, epochs, x_inicial)

print("El valor de x que minimiza la función es", result)
print("El valor de mínimo de la función es", fx(result))

plt.subplot(1,2,1)
plt.plot(iterations, y)
plt.xlabel("Iteraciones",fontsize=16, fontweight="bold")
plt.ylabel("y",fontsize=16, fontweight="bold")

X = np.linspace(-5,5,100)
Y = list(map(lambda x: fx(x), X))

# Graficamos
plt.subplot(1,2,2)
plt.plot(X,Y,result,fx(result),'ro')
plt.xlabel('x',fontsize=16, fontweight="bold")
plt.ylabel('y',fontsize=16, fontweight="bold")
plt.suptitle("Iteraciones y f(x)", fontsize=16, fontweight="bold")
plt.show()
`````
